```python
import numpy as np 
import pandas as pd 
from mlxtend.frequent_patterns import apriori, association_rules 
```

```python
cust_data = pd.read_csv(r"C:\Users\Predator\Downloads\Groceries data.csv") 
cust_data.head()
```

```python
# Step 1: Keep only necessary columns (Member_number and itemDescription)
cust_data = cust_data[['Member_number', 'itemDescription']]
```

```python
# Step 2: Group items by customer (transactions format)
transactions = cust_data.groupby('Member_number')['itemDescription'].apply(list)
```

```python
# Convert to list of lists for Apriori algorithm
transactions_list = transactions.tolist()
```

```python
# Display first few transactions
transactions_list[:5]
```

```python
from mlxtend.frequent_patterns import apriori, association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
# Step 3: Convert transactions into One-Hot Encoding format
te = TransactionEncoder()
te_ary = te.fit(transactions_list).transform(transactions_list)
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
# Step 4: Apply Apriori algorithm
frequent_itemsets = apriori(df_encoded, min_support=0.01, use_colnames=True)
```

```python
# Step 5: Generate Association Rules
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
```

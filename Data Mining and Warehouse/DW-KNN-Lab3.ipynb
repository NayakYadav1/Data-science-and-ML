```python
import pandas as pd
import numpy as np 
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv(r"C:\Users\Predator\Downloads\loan_data.csv")
```

```python
data
```

```python
data.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
categ_col = data.select_dtypes(include=['object']).columns
encoder = LabelEncoder()
```

```python
for col in categ_col:
    data[col] = encoder.fit_transform(data[col])
```

```python
data.dtypes
```

```python
X = data.drop('loan_status', axis = 1)
y = data['loan_status']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state=4, stratify = y)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model = KNeighborsClassifier()
```

```python
knn_model.fit(X_train, y_train)
```

```python
y_pred = knn_model.predict(X_test)
```

```python
y_pred
```

```python
y_test.values
```

```python
accuracy = accuracy_score(y_pred, y_test.values)
```

```python
accuracy
```

```python
conf = confusion_matrix(y_pred, y_test.values)
```

```python
conf
```

```python
class_rep = classification_report(y_pred, y_test.values)
```

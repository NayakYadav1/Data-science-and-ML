```python
# Importing Matplotlib
import matplotlib.pyplot as plt
```

```python
# Line Plots
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-label')
plt.title('Line Plot')
```

```python
# Scatter Plots
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.scatter(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
```

```python
# Bar Plots
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 5]

plt.bar(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot')
```

```python
# Barh Plots
categories = ['A', 'B', 'C', 'D']
values = [10, 20, 15, 5]

plt.barh(categories, values)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Barh Plot')
```

```python
# Histogram
import numpy as np
data = np.random.randn(1000)

plt.hist(data, bins = 30)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram Plot')
```

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
a = [2, 3, 5, 4, 5]
b = [2, 3, 7, 9, 10]
plt.plot(x, y, color = 'red', marker = 'o', label = 'Series 1')
plt.plot(a, b, linestyle = '--', label = 'Series 2')

plt.legend(loc = 'upper right')
```
